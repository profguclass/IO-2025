```python
import numpy as np   # Numpy import
import pandas as pd  # Pandas import
np.set_printoptions(precision=3, suppress=True)
```

3부문 총, 국산, 수입거래표 및 고용표 불러오기

```python
url_Tt_33 = "https://raw.githubusercontent.com/profguclass/IO-2025/main/Total_matrix_33sectors.csv"  # 총거래표(3부문) 불러오기
Tt_33 = pd.read_csv(url_Tt_33, thousands=',')

url_Dt_33 = "https://raw.githubusercontent.com/profguclass/IO-2025/main/Domestic_matrix_33sectors.csv"  # 국산거래표(3부문) 불러오기
Dt_33 = pd.read_csv(url_Dt_33, thousands=',')

url_It_33 = "https://raw.githubusercontent.com/profguclass/IO-2025/main/Imported_matrix_33sectors.csv"  # 수입거래표(3부문) 불러오기
It_33 = pd.read_csv(url_It_33, thousands=',')
```

```python
Tt = Tt_33.to_numpy()  # numpy로 불러오기
Dt = Dt_33.to_numpy()
It = It_33.to_numpy()

Tt
```

```python
Tt_1 = Tt[0:40,1:50].astype(float)
Dm_1 = Dt[0:33,1:50].astype(float)
Im_1 = It[0:33,1:50].astype(float)
```

```python
Ct = Tt_1[-1,0:33]  # 총산출액 벡터
Va = Tt_1[-2,0:33]  # 부가가치 벡터
```

```python
Ct_m = np.tile(Ct, reps=[33,1]) # 총산출액 벡터를 3번 행반복
```

```python
At = (Tt_1[0:33,0:33]/Ct_m)   # 총투입계수
Ad = (Dm_1[0:33,0:33]/Ct_m)   # 국산투입계수
Ai = (Im_1[0:33,0:33]/Ct_m)   # 수입투입계수
Av = (Va/Ct).reshape([1,33])  # 부가가치율 벡터

print('총투입계수표 \n', At.round(3),'\n')
print('국산투입계수표 \n', Ad.round(3),'\n')
print('수입투입계수표 \n', Ai.round(3),'\n')
print('부가가치율 \n', Av.round(3))

```

```python
df_At = pd.DataFrame(At)  #pandas로 내보내기
df_Ad = pd.DataFrame(Ad)  #pandas로 내보내기
df_Ai = pd.DataFrame(Ai)  #pandas로 내보내기
df_Av = pd.DataFrame(Av)                  #pandas로 내보내기
```

##### 생산유발계수(Lf)

```python
Lf = np.linalg.inv(np.identity(33)-Ad)  # 생산유발계수

Lf_rw_sm = Lf.sum(axis=0)  # 생산유발계수의 행합
Lf_cl_sm = Lf.sum(axis=1)  # 생산유발계수의 열합

Lf_bwd = (Lf.sum(axis=0)/Lf_rw_sm.mean()).astype(float).reshape([1,33])  #영향력 계수
Lf_fwd = (Lf.sum(axis=1)/Lf_cl_sm.mean()).astype(float).reshape([1,33])  #감응도 계수

print('생산유발계수 \n', Lf.round(3),'\n')

print('생산유발계수의 행합 \n', Lf_rw_sm.round(3),'\n')
print('생산유발계수의 열합 \n', Lf_cl_sm.round(3),'\n')

print('부문별 영향력계수 \n', Lf_bwd.round(3),'\n')
print('부문별 감응도계수 \n', Lf_fwd.round(3),'\n')
```

```python
df_Lf = pd.DataFrame(Lf.round(3))  #pandas로 내보내기
```

#### 수입유발계수(Li)

```python
Li = (Ai@Lf).round(3)  #수입유발계수
print('수입유발계수 \n', Li)
```

```python
df_Li = pd.DataFrame(Li.round(3))  #pandas로 내보내기
```

#### 부가가치유발계수(Lv)

```python
df_Lv = pd.DataFrame(Lv.round(3))  # pandas로 내보내기
```

```python
# CSV파일로 결과 내보내기
df_At.to_csv('총투입계수.csv', encoding='utf_8-sig')
df_Ad.to_csv('국산투입계수.csv', encoding='utf_8-sig')
df_Ai.to_csv('수입투입계수.csv', encoding='utf_8-sig')

df_Lf.to_csv('생산유발계수.csv', encoding='utf_8-sig')
df_Lv.to_csv('부가가치유발계수.csv', encoding='utf_8-sig')
df_Li.to_csv('수입유발계수.csv', encoding='utf_8-sig')
```
